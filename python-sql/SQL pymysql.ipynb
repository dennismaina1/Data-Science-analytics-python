```python
schools=pd.read_sql_query('SELECT * FROM chicagopublicschools',conn)
```

```python
#which schools in community area 10-15 are healthy school certified

#python
healthy = schools[schools.COMMUNITY_AREA_NUMBER.between(10,15) ]
healthy = healthy[healthy.HEALTHY_SCHOOL_CERTIFIED =='Yes']
healthy

#SQL
healthy2= pd.read_sql_query("select * from chicagopublicschools where HEALTHY_SCHOOL_CERTIFIED ='Yes'and COMMUNITY_AREA_NUMBER between 10 and 15",conn)
healthy2
```

```python
#average school safety score

#python
safety=schools.SAFETY_SCORE.mean()

#SQL
safety2=pd.read_sql_query("select AVG(SAFETY_SCORE) as mean_score from chicagopublicschools",conn)
safety2
```

```python
#top5 community areas by college enrollement

#python
toparea=schools.COLLEGE_ENROLLMENT.nlargest(n=5)
toparea2=schools[['COMMUNITY_AREA_NAME','COLLEGE_ENROLLMENT']].where(schools.COLLEGE_ENROLLMENT>=min(toparea))
toparea2=toparea2.dropna()
toparea2.sort_values('COLLEGE_ENROLLMENT',ascending=False)


#SQL
topareas= pd.read_sql_query('SELECT COMMUNITY_AREA_NAME,COLLEGE_ENROLLMENT from chicagopublicschools ORDER BY COLLEGE_ENROLLMENT DESC LIMIT 5 ',conn)
topareas
```

```python
#percapita income of area with score safety of 1

#python 
Per_capita= schools.COMMUNITY_AREA_NUMBER[schools.SAFETY_SCORE==1]
Per_capita= census.PER_CAPITA_INCOME.where(census.COMMUNITY_AREA_NUMBER==int(Per_capita)).dropna()
Per_capita

#SQL
Percapita=pd.read_sql_query("Select chicagopublicschools.SAFETY_SCORE,chicagopublicschools.COMMUNITY_AREA_NUMBER,PER_CAPITA_INCOME from chicagocensusdata,chicagopublicschools where \
                                chicagocensusdata.COMMUNITY_AREA_NUMBER=chicagopublicschools.COMMUNITY_AREA_NUMBER\
                                    AND SAFETY_SCORE='1'",conn)
Percapita
```
